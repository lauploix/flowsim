```python
import numpy as np
from scipy import signal

rng = np.random.default_rng()

raw = np.array ([0.0 if t < 50 else 1.0 for t in range(100)])

raw += [np.sin(t * 2 * np.pi / 50) for t in range(len(raw))]
raw += rng.standard_normal(len(raw))

x = np.concatenate([rng.standard_normal(500), raw + rng.standard_normal(len(raw)), rng.standard_normal(600) + [np.sin(t * 2 * np.pi / 50) for t in range(600)]])
#x  += rng.standard_normal(len(x)) # add noise

y = np.concatenate([raw])
#y  += rng.standard_normal(len(y)) # add noise

correlation = signal.correlate(x, y, mode="valid")
lags = signal.correlation_lags(x.size, y.size, mode="valid")

lag = lags[np.argmax(correlation)]

import matplotlib.pyplot as plt  

plt.plot(range(len(raw)), raw)  
plt.grid(True)  
plt.show()

plt.plot(range(len(x)), x)  
plt.plot(range(len(y)), y)  
plt.grid(True)  
plt.show()


plt.plot(lags, correlation)  
plt.grid(True)  
plt.show()

print (f"Y is ahedad of X by {lag} samples")
```
